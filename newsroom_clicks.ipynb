```python
df2 = df.rename(columns={'Column1':'publication_id','_1' :'publication label', 'Publication_Label' : 'Message Send_Date'})

#try doing this with the lambda and apply  function 
```

```python
df2["Message_Subject"] = df2[["Message_Subject", "Publication_ID"]].apply(lambda x:"".join(x),axis = 1)

#lambda in pandas can by applied to both columns and rows in a pandas dataframe 
#used ro make anonymous functions
#apply and lambda creates a new column whose values are dependent on values of an existing column
```

```python
nan_count = df2.isna().sum().sum()
null_count = df2.isnull().sum().sum()

sum_missing = nan_count + null_count

sum_missing
```

```python

df2.drop(['Publication_ID'],axis =1, inplace = True)
#axis 1 = the column,axis 0 = rows 
```

```python
df2['Message_Subject'] = df2['Message_Subject'].str.replace('{','',regex=True)
# regular expression which helps you search for an exact match in a string (REGEX)
```
